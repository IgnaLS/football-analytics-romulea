```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from mplsoccer import Pitch
from mplsoccer import VerticalPitch
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
from matplotlib.patches import Patch
import seaborn as sns
```

```python
# Importing Datasets
df_1 = pd.read_csv('Romulea_vs_Ultima.csv')
df_2 = pd.read_csv('Romulea_vs_Pomezia.csv')
df_3 = pd.read_csv('Romulea_vs_Boreale.csv')
df_4 = pd.read_csv('Romulea_vs_W3 Maccarese.csv')
df_5 = pd.read_csv('Romulea_vs_Roma City.csv')
df_6 = pd.read_csv('Romulea_vs_Civitavecchia.csv')
df_7 = pd.read_csv('Romulea_vs_Viterbese.csv')
```

```python
df_1['Avversario'] = 'Ultima'
df_2['Avversario'] = 'Pomezia'
df_3['Avversario'] = 'Boreale'
df_4['Avversario'] = 'W3 Maccarese'
df_5['Avversario'] = 'Roma City'
df_6['Avversario'] = 'Civitavecchia'
df_7['Avversario'] = 'Viterbese'
```

```python
df_1
```

```python
df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7], ignore_index=True)
```

```python
df.head(50)
```

```python
df.info()
```

```python
# Cambiare tipo di dato a X2 e Y2

df[['X2', 'Y2']] = df[['X2', 'Y2']].replace('-', np.nan)
df[['X2', 'Y2']] = df[['X2', 'Y2']].astype(float)
```

```python
df.to_csv('Romulea_casa.csv')
```

```python
# Inserire colonna 'Goal'
df['Goal/NoGoal'] = ''

# Riempire se Goal
conditions = [
    df['Avversario'] == 'Ultima',
    df['Event'] == 'Goal',
    (df['X2'].notna()) & (df['Event'] == 'Shot') & (df['X2'] != 92)
    ]

values = ['', 'Goal', 'Goal']

df['Goal/NoGoal'] = np.select(conditions, values, default = '')
```

```python
df[df['Goal/NoGoal'] == 'Goal']
```

```python
pitch = Pitch(pitch_type= 'statsbomb')

print(f"Lunghezza: {pitch.dim.pitch_length}")
print(f"Larghezza: {pitch.dim.pitch_width}")
print(f"Extent: {pitch.dim.pitch_extent}")
```

```python
df['X_stab'] = df['X'] * 1.2
df['X2_stab'] = df['X2'] * 1.2
df['Y_stab'] = df['Y'] * 0.8
df['Y2_stab'] = df['Y2'] * 0.8
```

```python
ver_pitch = VerticalPitch(pitch_type='statsbomb')
print(f"Lunghezza: {ver_pitch.dim.pitch_length}")
print(f"Larghezza: {ver_pitch.dim.pitch_width}")
print(f"Extent: {ver_pitch.dim.pitch_extent}")
```

```python
#subset for Romulea's Shots
df_romulea = df[(df['Team'] == 'Romulea') & (df['Event'].isin(['Shot', 'Goal']))]
df_romulea = df_romulea.reset_index(drop=True)
df_romulea
```

```python
# Andare a invertire le y
conditions = [
    df_romulea['X_stab'] < 60
    ]

values = [80 - df_romulea['Y_stab']]

df_romulea['Y_stab'] = np.select(conditions, values, default = df_romulea['Y_stab'])

df_romulea['Y_stab'] = df_romulea['Y_stab'].round(2)
```

```python
# Invertire X per portarli tutti da una parte
conditions = [
    df_romulea['X_stab'] < 60
    ]

values = [120 - df_romulea['X_stab']]

df_romulea['X_stab'] = np.select(conditions, values, default = df_romulea['X_stab'])

df_romulea['X_stab'] = df_romulea['X_stab'].round(2)
```

```python
# Calcolo tiri totali e goal
tiri = df_romulea.shape[0]
goals = df_romulea[df_romulea['Goal/NoGoal'] == 'Goal'].shape[0]
print(tiri)
print(goals)
```

```python
# Calcolo distanza media
av_dist = (120 - df_romulea['X_stab'].mean()).round(2)
act_dist = (av_dist * (105/120)).round(2)
ave = 120 - av_dist
print(av_dist)
print(act_dist)
print(ave)
```

```python
# Subset for opponent's shots
df_adv = df[(df['Team'] != 'Romulea') & (df['Event'].isin(['Shot', 'Goal'])) & (df['Y'] != 2)]
df_adv = df_adv.reset_index(drop=True)
df_adv
```

```python
# Andare a invertire le y
conditions = [
    df_adv['X_stab'] < 60
    ]

values = [80 - df_adv['Y_stab']]

df_adv['Y_stab'] = np.select(conditions, values, default = df_adv['Y_stab'])
df_adv['Y_stab'] = df_adv['Y_stab'].round(2)
```

```python
# Andare a invertire le x
conditions = [
    df_adv['X_stab'] < 60
    ]

values = [120 - df_adv['X_stab']]

df_adv['X_stab'] = np.select(conditions, values, default = df_adv['X_stab'])
df_adv['X_stab'] = df_adv['X_stab'].round(2)
```

# Analisi Passaggi

```python
# Subset Passaggi Romulea
df_rom_pass = df[(df['Team'] == 'Romulea') & (df['Event'].isin(['Pass', 'Shot Assist', 'Cross']))]
df_rom_pass = df_rom_pass.reset_index(drop=True)
df_rom_pass
```

```python
# Plotting
fig, ax = plt.subplots(figsize = (8, 13.5))
fig.set_facecolor(background)
ax.patch.set_facecolor(background)

pitch = Pitch(
    pitch_type='statsbomb',
    half=False,
    pitch_color=background,
    pad_bottom=.5,
    line_color='white',
    linewidth=.75,
    axis=True,
    label=True
)

for x in range(len(df_rom_pass['X_stab'])):
    if df_rom_pass['Event'][x] == 'Shot Assist':
    # plt.plot((df_rom_pass['X'][x], df_rom_pass['X2'][x]), (df_rom_pass['Y'][x], df_rom_pass['Y2'][x]), color = 'red')
        pitch.arrows(df_rom_pass['X_stab'][x], df_rom_pass['Y_stab'][x], df_rom_pass['X2_stab'][x], df_rom_pass['Y2_stab'][x], ax=ax, color='red',  width=2, headwidth=5, headlength=5, alpha=0.7)
    # plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color = 'red')
    if df_rom_pass['Event'][x] == 'Pass':
        pitch.arrows(df_rom_pass['X_stab'][x], df_rom_pass['Y_stab'][x], df_rom_pass['X2_stab'][x], df_rom_pass['Y2_stab'][x], ax=ax, color='green',  width=2, headwidth=5, headlength=5, alpha=0.7)
        #plt.scatter(df['endX'][x], df['endY'][x], color = 'red')
    if df_rom_pass['Event'][x] == 'Cross':
        pitch.arrows(df_rom_pass['X_stab'][x], df_rom_pass['Y_stab'][x], df_rom_pass['X2_stab'][x], df_rom_pass['Y2_stab'][x], ax=ax, color='yellow',  width=2, headwidth=5, headlength=5, alpha=0.7)

pitch.draw(ax = ax)
# plt.gca().invert_yaxis()

# Crea legenda
legend_elements = [
    Patch(facecolor='red', edgecolor='white', label='Assist'),
    Patch(facecolor='green', edgecolor='white', label='Passaggi'),
    Patch(facecolor='yellow', edgecolor='white', label='Cross')
]

ax.legend(handles=legend_elements, loc='lower right', facecolor=background, 
          edgecolor='white', labelcolor='white', fontsize=11)

plt.tight_layout()
plt.title('Passaggi Romulea', color = 'white', fontsize = 20)
plt.show()

```

```python
# Plotting Pass Heatmap 
fig, ax = plt.subplots(figsize=(8, 13.5))
fig.set_facecolor(background)
ax.patch.set_facecolor(background)

pitch = Pitch(
    pitch_type='statsbomb',
    half=False,
    pitch_color=background,
    pad_bottom=.5,
    line_color='white',
    linewidth=.75,
    axis=True,
    label=True
)

# Disegna il campo
pitch.draw(ax=ax)

# HEATMAP dei passaggi (zone di partenza)
kde = pitch.kdeplot(
    df_rom_pass['X_stab'],
    df_rom_pass['Y_stab'],
    ax=ax,
    cmap='hot',
    shade=True,
    levels=100,
    alpha=0.7
)
plt.title('Inizio Passaggi Romulea (Heatmap)', color = 'white', fontsize = 20)
```
